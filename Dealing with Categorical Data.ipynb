```python
df['MSSubClass'] = df['MSSubClass'].apply(str)
```

```python
my_object_df = df.select_dtypes(include='object') ## Selecting all string value columns
```

```python
my_numeric_df = df.select_dtypes(exclude='object')
```

```python
my_object_dummies = pd.get_dummies(my_object_df,drop_first=True)
```

```python
my_object_dummies
```

```python
## Let's join Numeric and Object Dummies using concat function
final_df = pd.concat([my_object_dummies , my_numeric_df], axis=1)
```
